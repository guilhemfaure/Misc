# Parse a PDB with Biopython

### Import library and Initialize parser 

```
from Bio.PDB import *
parser_pdb = PDBParser()
```

### Load PDB

```
Mypdb = parser_pdb.get_structure('mystructure', '1a2k.pdb')
```

### Get Coord Calpha

```
# This print Res Number and XYZ
for model in Mypdb:
    for chain in model:
        for residue in chain:
            res_name = residue.get_resname()
            ca_coord = list(residue['CA'].get_coord()) if residue.has_id('CA') else [None]*3
            res_nb = residue.get_id()
            print ('{res}\t{nb}\t{coords}'.
                   format(res = res_name, 
                          nb  = res_nb[1], 
                          coords = ' '.join(map(str, ca_coord)))
                   )
```
